```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import datetime
```

```python
index = '^GSPC'
index_name = 'S&P 500'

# These date variables are for the graphs to specify in the title the time period
# Start_date_files is the variable used in the file name since file names can't have /
start_date = datetime.date(2000,1,1).strftime('%m/%d/%Y')
start_date_files = datetime.date(2000,1,1).strftime('%m.%d.%Y')

end_date = datetime.datetime.today().strftime('%m/%d/%Y')
end_date_files = datetime.datetime.today().strftime('%m.%d.%Y')

years = range(2000,2021)
years_datetime = []

for year in years:
    years_datetime.append(datetime.datetime(year,1,1))
```

```python
data = pd.DataFrame(index=range(255))

for year in years:
    print(year)
    stock_performance = wb.DataReader(index, data_source='yahoo', start=f'{year-1}-12-31', end=f'{year}-12-31')['Adj Close']
    stock_performance.to_csv(f'/Users/matthulley/Documents/Coding/Python/Jupyter Projects/S&P 500 {year} Performance.csv',header=False)
    csv_file = pd.read_csv(f'/Users/matthulley/Documents/Coding/Python/Jupyter Projects/S&P 500 {year} Performance.csv', header=None)
    data[year] = csv_file[1]
```

```python
chart = plt.figure(figsize=(15,10))
#plt.plot(data / data.iloc[0] - 1)
plt.plot(data / data.iloc[0] - 1, alpha=0.7, color='grey' )
plt.plot(data[2001] / data[2001].iloc[0] - 1, alpha=1.0, color='green', label='2001')
plt.plot(data[2002] / data[2002].iloc[0] - 1, alpha=1.0, color='purple', label='2002')
plt.plot(data[2008] / data[2008].iloc[0] - 1, alpha=1.0, color='red', label='2008')
plt.plot(data[2009] / data[2009].iloc[0] - 1, alpha=1.0, color='darkorange', label='2009')
plt.plot(data[2020] / data[2020].iloc[0] - 1, alpha=1.0, color='blue', label='2020')

plt.minorticks_on()
plt.grid(True, which='major', axis='both')
plt.grid(True, which='minor', axis='both', linestyle='--', alpha=0.5)

plt.title(f'Historical {index_name} Performance {start_date} - {end_date}')
plt.ylabel('Performance')
plt.xlabel('Days into the Year')

plt.legend()
#plt.legend(data.columns.values)
plt.show()
#chart.savefig(f'/Users/matthulley/Documents/Coding/Charts/Historical {index_name} Performance {start_date_files} - {end_date_files} (Grey).pdf')
```

```python
trough = []
peak = []
final = []
start = []
number_of_empty = data.isna().sum()

for year in years:
    trough.append(data[year].min())
    peak.append(data[year].max())
    final.append(data[year].iloc[-(number_of_empty[year]+1)])
    start.append(data[year].iloc[0])
```

```python
chart = plt.figure(figsize=(15,10))
plt.scatter(years_datetime, start, label='start')
plt.scatter(years_datetime, peak, label='peak')
plt.scatter(years_datetime, trough, label='trough')
plt.plot(years_datetime, final, label='final')

plt.minorticks_on()
plt.grid(True, which='major', axis='both')
plt.grid(True, which='minor', axis='both', linestyle='--', alpha=0.5)

plt.title(f'{index_name} Performance {start_date} - {end_date}')
plt.ylabel(f'{index_name} Index Level')
plt.xlabel('Years')

plt.legend()
plt.show()
#chart.savefig(f'/Users/matthulley/Documents/Coding/Charts/{index_name} Index Level {start_date_files} - {end_date_files}.pdf')
```

```python
trough_percent = []
peak_percent = []
final_percent = []
start_percent = []

for year in years:
    trough_percent.append(data[year].min()/data[year].iloc[0] - 1)
    peak_percent.append(data[year].max()/data[year].iloc[0] - 1)
    final_percent.append(data[year].iloc[-(number_of_empty[year]+1)]/data[year].iloc[0] - 1)
    start_percent.append(data[year].iloc[0]/data[year].iloc[0] - 1)
```

```python
chart_percent = plt.figure(figsize=(15,10))

plt.scatter(years_datetime, start_percent, label='start')
plt.scatter(years_datetime, peak_percent, label='peak')
plt.scatter(years_datetime, trough_percent, label='trough')
plt.plot(years_datetime, final_percent, label='final')

plt.title(f'{index_name} Performance {start_date} - {end_date}')
plt.ylabel(f'{index_name} Performance')
plt.xlabel('Years')

plt.minorticks_on()
plt.grid(True, which='major', axis='both')
plt.grid(True, which='minor', axis='both', linestyle='--', alpha=0.5)

plt.legend()
plt.show()
#chart_percent.savefig(f'/Users/matthulley/Documents/Coding/Charts/{index_name} Returns {start_date_files} - {end_date_files}.pdf')
```
